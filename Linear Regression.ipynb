```python
import numpy as np

x = np.arange(200) * .5
delta = np.random.normal(size=(200,))
y = x * .05  + 3 + delta
```

```python
import matplotlib.pyplot as plt
%matplotlib inline  

plt.plot(x,y)
```

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
features = x.reshape(-1,1)
lin_reg.fit(features,y)
```

```python
best_fit = lin_reg.predict(features)
plt.plot(x,best_fit)
plt.scatter(x,y)
```
